Se empleará este algoritmo de agrupamiento para descubrir patrones o segmentos con características similares con respecto al nivel de satisfacción. Este es un algoritmo de aprendizaje no supervisado, el cual consta básicamente de centroides y es el algoritmo de agrupamiento más popular, su función consiste en minimizar la varianza de los datos (Betancourt, 2021). A continuación, se describe como se implementará este algoritmo:

• Se inicia importando el algoritmo.

• Determinar el número óptimo de clusters.

• Definir el modelo K-means.

**Random Forest.**

Se empleará este algoritmo para predecir el nivel de satisfacción del pasajero. Este es un algoritmo de aprendizaje supervisado, siendo un algoritmo popular para resolver problemas de clasificación y regresión. El funcionamiento de este consta de la combinación de cientos de arboles de decisión para posteriormente entrenar cada árbol de decisión, una vez echo esto, las predicciones finales del algoritmo se realizan utilizando el promedio de las predicciones de cada árbol de forma individual (Cardellino, 2021). A continuación, se describe como se implementará este algoritmo:

• Se inicia importando el algoritmo.

• Se divide el conjunto de datos, la variable objetivo se denomina como “satisfaction”.

• Se evalúa el rendimiento del modelo para realizar las predicciones por medio de métricas como precisión, Recall y F1-Score.

## <p style="background-color:Blue;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">10. Modelo de aprendizaje no supervisado</p>

```python
# Crear copia del dataset
dataframe_n = dataframe.copy()
# Normalizar los datos
scaler = StandardScaler()
dataframe_scaled = scaler.fit_transform(dataframe_n)
```

```python
# Implementando PCA
# instancia (variable) para la función PCA (definir número de componentes principales a reducir)
pca = PCA(n_components=3)
# aplicar PCA
pca.fit(dataframe_scaled)
# Crear dataframe del PCA implementado
PCA_dataframe = pd.DataFrame(pca.transform(dataframe_scaled), columns=(["col1","col2", "col3"]))
# Mostrar estadísticas del dataframe creado con el PCA
PCA_dataframe.describe().T
```

```python
# Lista para almacenar los valores de la inercia
inertia = []

# Valores de k que queremos probar
k_values = range(1, 11)

# Entrenar modelos K-means para cada valor de k y calcular la inercia
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(PCA_dataframe)
    inertia.append(kmeans.inertia_)

# Graficar el método del codo
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del codo para determinar k')
plt.show()
```

De acuerdo al método del codo, el número de clusters es 4.

```python
# Crear el modelo K-means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(PCA_dataframe)
```

## <p style="background-color:Blue;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">11. Evaluación del modelo de aprendizaje no supervisado</p>

```python
# Obtener las etiquetas de los clusters
labels = kmeans.labels_

# Agregar las etiquetas de cluster al DataFrame original
dataframe_n['Cluster'] = labels

# Visualizar los clusters
plt.figure(figsize=[4,3]) # tamaño
pl = sns.countplot(x=dataframe_n["Cluster"]) # definir plot
pl.set_title("Distribución de Clusters") # título
plt.show()
```

```python
# Calculando la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusión:")
print(conf_matrix)

# Calculando la precisión
accuracy = accuracy_score(y_test, y_pred)
print("Precisión:", accuracy)

# Calculando la precisión por clase
precision = precision_score(y_test, y_pred, average=None)
print("Precisión por Clase:", precision)

# Calculando el recall por clase
recall = recall_score(y_test, y_pred, average=None)
print("Recall por Clase:", recall)

# Calculando el F1-score por clase
f1 = f1_score(y_test, y_pred, average=None)
print("F1-Score por Clase:", f1)
```

## <p style="background-color:Blue;font-family:newtimeroman;color:#FFF9ED;font-size:150%;text-align:center;border-radius:10px 10px;">14. Optimización del modelo</p>

```python
from sklearn.model_selection import GridSearchCV

# Definir los hiperparámetros a ajustar
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Inicializar el RandomForestClassifier
rf_model = RandomForestClassifier(random_state=0)

# Realizar la búsqueda en cuadrícula utilizando validación cruzada
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)

# Obtener los mejores hiperparámetros
best_params = grid_search.best_params_
print("Mejores hiperparámetros:", best_params)

# Entrenar el modelo con los mejores hiperparámetros
best_rf_model = RandomForestClassifier(random_state=0, **best_params)
best_rf_model.fit(x_train, y_train)

# Evaluar el modelo optimizado
y_pred = best_rf_model.predict(x_test)
print("Precisión del modelo optimizado:", accuracy_score(y_test, y_pred))
```
